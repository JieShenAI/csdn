```python
#Model Download
from modelscope import snapshot_download

model_dir = snapshot_download('AI-ModelScope/bert-base-uncased')
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

bert_embedding_hf = HuggingFaceEmbeddings(model_name=model_dir)
```

```python
hf_encode = bert_embedding_hf.embed_documents(["hello new world!"])
```

```python
hf_encode_tensor = torch.tensor(hf_encode)
```

```python
hf_encode_tensor
```

```python
bi_encoder = BiEncoderModel(model_name=model_dir, sentence_pooling_method="mean")
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_dir)
```

```python
bi_encoder_tensor = bi_encoder.encode(tokenizer(["hello new world!"], return_tensors="pt"))
```

```python
sum(sum(bi_encoder_tensor - hf_encode_tensor))
```

```python
bi_encoder_cls = BiEncoderModel(model_name=model_dir, sentence_pooling_method="cls")
```

```python
bi_encode_cls_tensor = bi_encoder_cls.encode(
    tokenizer(["hello new world!"], return_tensors="pt")
)
```
