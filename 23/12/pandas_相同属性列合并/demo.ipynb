```python
import os

import pandas as pd
from pandas import DataFrame, read_csv
filename = "data/data.csv"
```

```python
head = read_csv(filename, nrows=10)
```

```python
keys = ['t', 'i', 'j']
attrs = ['v']

def init_df(filename):
    # data = read_csv(filename, nrows=10) # 读取少量数据，方便查看结果
    data = read_csv(filename)
    delete_col = set(data.columns.values) - set(keys) - set(attrs)
    data = data.drop(list(delete_col), axis=1)

    d = {key: [] for key in keys + attrs}
    return data, pd.DataFrame.from_dict(d)
```

```python
data_df, new_df = init_df(filename)
```

```python
def is_equal(a: DataFrame, b: DataFrame) -> bool:
    for key in keys:
        if a[key] != b[key]:
            return False
    return True
```

```python
(
    is_equal(data_df.iloc[0], data_df.iloc[0]), 
    is_equal(data_df.iloc[0], data_df.iloc[-1])
)
```

如何相加

```python
def add(a, b):
    assert is_equal(a, b)
    for attr in attrs:
        a[attr] += b[attr]
```

```python
def insert(df:DataFrame, other:DataFrame):
    df.loc[len(df)] = other
```

```python
def main(df, res):
    insert(res, df.iloc[0])
    for idx in range(1, len(df)):
        item = df.iloc[idx]
        if is_equal(item, res.iloc[-1]):
            add(res.iloc[-1], item,)
        else:
            insert(res, item)
```

```python
main(data_df, new_df)
```
